```python
#ler o arquivo csv na pasta arquivo_csv
df = pd.read_csv("arquivos_csv/combustiveis.csv")
```

```python
df.head()
```

```python
#pegar somente os dados do estado de Distrito Federal
df = df[df['sigla_uf']=='DF']
#pegar os dados somente do produto gasolina
df = df[df['produto']=='gasolina']
#resetar o index
df = df.reset_index(drop=True)
```

```python
df
```

```python
#transformar a coluna data_coleta em datetime
df['data_coleta'] = pd.to_datetime(df['data_coleta'], format='%Y-%m-%d')
```

```python

df['mes_ano'] = df['data_coleta'].dt.strftime('%m/%Y')
```

```python
df
```

```python
#verificar valores nulos
df.isnull().sum()
```

```python
#criar dataframe como mes_ano e preco de venda
df_preco = df[['mes_ano','preco_venda']]
```

```python
#criar uma coluna somene com ano
df_preco['ano'] = df_preco['mes_ano'].str[-4:]
```

```python
#ordenar o dataframe pelo ano
df_preco = df_preco.sort_values(by='ano')
```

```python
df_preco
```

```python
#resetar o index
df_preco = df_preco.reset_index(drop=True)
```

```python
df_preco[20:30]
```

```python
#criar um dataframe com o ano e o preço médio
df_preco_medio = df_preco.groupby('mes_ano')['preco_venda'].mean()
```

```python
#mudar o tipo do mes_ano para datetime
df_preco_medio.index = pd.to_datetime(df_preco_medio.index, format='%m/%Y')
```

```python
#ordenar o dataframe pelo ano
df_preco_medio = df_preco_medio.sort_index()
```

```python
#remover o dia da data
df_preco_medio.index = df_preco_medio.index.strftime('%m/%Y')
```

```python
#arredondar os valores para 2 casas decimais
df_preco_medio = df_preco_medio.round(2)
```

```python
df_preco_medio
```

```python
#salvar o dataframe em um arquivo excel
df_preco_medio.to_excel('arquivos_csv/preco_medio_gasolina.xlsx')
```
