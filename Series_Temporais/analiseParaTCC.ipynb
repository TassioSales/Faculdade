```python
import pandas as pd
import gc
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#ler o csv na pasta arquivos_csv

df = pd.read_csv('arquivos_csv/combustiveis.csv')
df.head() 
```

```python
#verificar o nome das colunas
df.columns
```

```python
colunas = ['ano', 'sigla_uf','municipio', 'cep_revenda', 'endereco_revenda', 'cnpj_revenda','nome_estabelecimento', 'bandeira_revenda', 'data_coleta', 'produto',
       'unidade_medida', 'preco_compra', 'preco_venda']
#colocar dataframe na ordem das colunas
df = df[colunas]
```

```python
df.head()
```

```python
#pegar valores unicos de uma coluna produto
df['produto'].unique()
```

```python
produtos_necessarios = ['gnv', 'diesel', 'etanol', 'gasolina']
#deixar apenas os produtos necessarios
df = df[df['produto'].isin(produtos_necessarios)]
#colocar nome dos produtos em maiusculo
df['produto'] = df['produto'].str.upper()
#substituir GASOLINA por GASOLINA COMUM
df['produto'] = df['produto'].str.replace('GASOLINA', 'GASOLINA COMUM')
#substituir ETANOL por ETANOL HIDRATADO
df['produto'] = df['produto'].str.replace('ETANOL', 'ETANOL HIDRATADO')
```

```python
#Verificar Distribuição de frequência dos dados
df['produto'].value_counts()
```

```python
df_describe = df.describe()
df_describe.T
```

```python
#Criar tabela com Frequencia, Percentual, Cumulativa Frequencia e Cumulativa Percentual
tabela = pd.crosstab(index=df['produto'], columns='count')
tabela['percentual'] = tabela['count']/tabela['count'].sum()
tabela['cumulativa_frequencia'] = tabela['count'].cumsum()
tabela['cumulativa_percentual'] = tabela['cumulativa_frequencia']/tabela['count'].sum()
#tranformar coluna produto em index
tabela = tabela.reset_index()
#renomear coluna count para frequencia
tabela.rename(columns={'count':'frequencia'}, inplace=True)
#transformar percentual e cumulativa percentual em porcentagem
tabela['percentual'] = tabela['percentual'].apply(lambda x: '{:.2%}'.format(x))
tabela['cumulativa_percentual'] = tabela['cumulativa_percentual'].apply(lambda x: '{:.2%}'.format(x))
#deixa a tabela tabela sem index
tabela.set_index('produto', inplace=True)
tabela.index.name = None
#renomear colunas col_0 para produto
tabela.rename_axis('Produto', axis=1, inplace=True)
#mostrar tabela na cor branca
tabela.style.background_gradient(cmap='Blues')
#colocar nome das colunas em Capitalize
tabela.columns = tabela.columns.str.capitalize()
tabela
```

```python
def cria_freq_percentual(coluna):
    #Criar tabela com Frequencia, Percentual, Cumulativa Frequencia e Cumulativa Percentual da GASOLINA COMUM por ano
    tabela = pd.crosstab(index=df[df['produto'] == coluna]['ano'], columns='count')
    tabela['percentual'] = tabela['count']/tabela['count'].sum()
    tabela['cumulativa_frequencia'] = tabela['count'].cumsum()
    tabela['cumulativa_percentual'] = tabela['cumulativa_frequencia']/tabela['count'].sum()
    #transformar percentual e cumulativa percentual em porcentagem
    tabela['percentual'] = tabela['percentual'].apply(lambda x: '{:.2%}'.format(x))
    tabela['cumulativa_percentual'] = tabela['cumulativa_percentual'].apply(lambda x: '{:.2%}'.format(x))
    tabela = tabela.reset_index()
    #deixa a tabela tabela sem index
    tabela.set_index('ano', inplace=True)
    tabela.index.name = None
    #renomear colunas col_0 para ano
    tabela.rename_axis('Ano', axis=1, inplace=True)
    #renomear coluna count para frequencia
    tabela.rename(columns={'count':'frequencia'}, inplace=True)
    #colocar nome das colunas em Capitalize
    tabela.columns = tabela.columns.str.capitalize()
    return tabela
```

```python
cria_freq_percentual('DIESEL')
```

```python
#separar produtos em dataframes diferentes
df_gasolina = df[df['produto'] == 'GASOLINA COMUM']
df_etanol = df[df['produto'] == 'ETANOL HIDRATADO']
df_gnv = df[df['produto'] == 'GNV']
df_glp = df[df['produto'] == 'GLP']
df_diesel = df[df['produto'] == 'DIESEL']
```

```python
#criar histograma com media do preco de compra da gasolina
def histograma_media(coluna, produto):
    media = coluna.mean()
    coluna.hist(bins=100, figsize=(10,5))
    plt.axvline(media, color='red', linestyle='dashed', linewidth=2)
    plt.title('Histograma de Preço de Compra de {}'.format(produto))
    plt.xlabel('Preço de Compra')
    plt.ylabel('Frequência')
    plt.show()
```

```python
histograma_media(df_gasolina['preco_venda'], 'Gasolina')
```

```python
histograma_media(df_etanol['preco_venda'], 'Etanol')
```

```python
histograma_media(df_gnv['preco_venda'], 'GNV')
```

```python
histograma_media(df_glp['preco_venda'], 'GLP')
```

```python
histograma_media(df_glp['preco_venda'], 'Diesel')
```

```python
#plotar boxplot do preco de venda da gasolina por ano
def boxplot_preco_venda(coluna, produto):
    plt.figure(figsize=(16,8))
    sns.boxplot(x='ano', y='preco_venda', data=coluna)
    plt.title('Boxplot de Preço de Venda de {}'.format(produto))
    plt.xlabel('Ano')
    plt.ylabel('Preço de Venda')
    #melhora a visualização do boxplot
    plt.ylim(0,8)
    plt.show()
```

```python
boxplot_preco_venda(df_gasolina, 'Gasolina')
```

```python
boxplot_preco_venda(df_etanol, 'Etanol')
```

```python
boxplot_preco_venda(df_gnv, 'GNV')
```

```python
boxplot_preco_venda(df_diesel, 'Diesel')
```
