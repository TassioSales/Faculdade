QUESTÃO 06) O que é backpropagation e como ele funciona? Qual a diferença entre
backpropagation e autodiff de modo reverso?

Resposta 6 - É um método computacionalmente eficiente para o treinamento de redes MLPs e que resolve o problema de realizar a propagação reversa do erro em RNAs com múltiplas camadas.

backpropagation refere-se a todo o processo de treinamento de uma rede neural artificial usando múltiplas etapas de retropagação, cada uma das quais computa gradientes e os usa para executar uma etapa de Descida gradiente. Em contraste, o diferencial automático de modo reverso é simplesmente uma técnica usada para calcular gradientes de forma eficiente e acontece de ser usada por retropagamento.
