```python
import numpy as np
```

```python
def hat_matrix(x: list) -> list:
    _denominador = np.dot(x.T, x)
    return x / _denominador
```

```python
if __name__ == '__main__':
    n = 5000
    x = np.array([np.random.rand() for _ in range(n)])
    noise = np.array([np.random.randn() for _ in range(n)])
    y = np.array(1.2 * x + noise)
    beta = hat_matrix(x)
    print(beta)
```

```python
def fit(beta: list, y: list) -> float:
    return np.dot(beta, y)
```

```python
model = fit(beta, y)
print(model)
```

```python
y_hat = np.array(model * x)
print(y_hat)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y, alpha = 0.2, label='Training data')
plt.plot(x, y_hat, color='brown', label=round(model, 2))
plt.legend()
plt.show()
```
