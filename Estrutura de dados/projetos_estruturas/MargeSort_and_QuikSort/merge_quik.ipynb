```python
def merge_sort(dados_para_merge):
    dados = dados_para_merge.sort_values(by="Age", kind="mergesort")
    return dados
```

```python
def quick_sort(dados_para_quick):
    dados = dados_para_quick.sort_values(by="Age", kind="quicksort")
    return dados
```

```python
inicio_merge = time.time()
merge_sort(df_filter)
fim_merge = time.time()
total_time_merge =  fim_merge - inicio_merge
merge = merge_sort(df_filter)
print(" #### DEPOIS DO MERGE SORT ####")
display(merge)
print(f"O fim_mergem de excução do MERGE SORT foi {total_time_merge}")
```
