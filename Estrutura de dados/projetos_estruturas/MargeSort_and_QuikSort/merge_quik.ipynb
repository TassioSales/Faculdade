<h1>Merge Sort</h1>
<p>MO Merge Sort (ordenaÃ§Ã£o por intercalaÃ§Ã£o/fusÃ£o) Ã© um algoritmo de ordenaÃ§Ã£o recursivo cujo mÃ©rito pela criaÃ§Ã£o Ã© atribuÃ­do ao fÃ­sico e matemÃ¡tico John von Neumann.</p>
<p>
    A ideia do Merge Sort Ã© dividir o vetor em dois subvetores, cada um com metade dos elementos do vetor original. Esse procedimento Ã© entÃ£o reaplicado aos dois subvetores recursivamente.
    Quando os subvetores tÃªm apenas um elemento (caso base), a recursÃ£o para. EntÃ£o, os subvetores ordenados sÃ£o fundidos (ou intercalados) num Ãºnico vetor ordenado.
</p>

```python
def merge_sort(dados_para_merge):
    dados = dados_para_merge.sort_values(by="Age", kind="mergesort")
    return dados
```

<h1>Quicksort </h1>

<p>
    Criado por Tony Hoare, o Quicksort (ordenaÃ§Ã£o rÃ¡pida) Ã© um algoritmo de ordenaÃ§Ã£o recursivo de divisÃ£o e conquista.
    Uma das suas vantagens Ã© que a ordenaÃ§Ã£o Ã© realizada no prÃ³prio vetor (in-place). Isso significa que nÃ£o hÃ¡ necessidade de alocar memÃ³ria auxiliar para realizar a ordenaÃ§Ã£o.
    Sua complexidade no tempo Ã© O(nlogn) no melhor caso e no caso mÃ©dio e O(n2) no pior caso. Contudo, em geral, o Quicksort apresenta um desempenho superior a outros algoritmos similares, como o Merge Sort.
</p>

<p>
    O Quicksort utiliza a seguinte lÃ³gica recursiva para ordenar um vetor
    Se o vetor sÃ³ possui um elemento (caso base):
    O vetor jÃ¡ estÃ¡ ordenado, retorne-o
    SenÃ£o
    Escolha um elemento do vetor, que chamaremos de pivÃ´;
    Divida o vetor em duas partiÃ§Ãµes:
    Uma partiÃ§Ã£o (Ã  esquerda) com os elementos menores ou iguais ao pivÃ´;
    Uma partiÃ§Ã£o (Ã  direita) com os elementos maiores que o pivÃ´;
    Reaplique o mÃ©todo nas duas partiÃ§Ãµes.
    Como nÃ£o existe uma regra para a escolha do pivÃ´, entÃ£o existem vÃ¡rias verÃµes do Quicksort. A versÃ£o original, de Hoare, escolhe o primeiro elemento do vetor como pivÃ´. Em Algoritmos: teoria e prÃ¡tica, Cormen [1] apresenta uma versÃ£o que utiliza o Ãºltimo elemento como pivÃ´.
    A versÃ£o de Hoare, em pseudocÃ³digo, Ã© dada a seguir
</p>

```python
def quick_sort(dados_para_quick):
    dados = dados_para_quick.sort_values(by="Age", kind="quicksort")
    return dados
```

<h1>Complexidade assintÃ³tica</h1>
<p>O melhor caso do Quicksort ocorre quando as partiÃ§Ãµes sÃ£o balanceadas, isto Ã©, cada partiÃ§Ã£o com n/2 elementos. A equaÃ§Ã£o de recorrÃªncia nesse caso Ã© igual a do Merge Sort</p>

<h2>T(n)=2T(n/2)+Î˜(n)(melhor caso)</h2>
<p>
    Î˜(n) Ã© a complexidade no tempo do mÃ©todo de partiÃ§Ã£o. A soluÃ§Ã£o dessa recorrÃªncia Ã© T(n)=Î˜(nlogn) ou, equivalentemente, T(n)=O(logn) e pode ser obtida atravÃ©s do teorema mestre ou pelo mÃ©todo de Akra-Bazzi.
    JÃ¡ o pior caso do Quicksort ocorre quando as chamadas recursivas produzem partiÃ§Ãµes com 0 e nâˆ’1 elementos. A partiÃ§Ã£o de tamanho zero fica Ã  direita ou Ã  esquerda do pivÃ´ (depende do vetor).
    Nas versÃµes de Hoare e Cormen, esse caso patolÃ³gico ocorre quando o array estÃ¡ ordenado em ordem crescente ou descrente. A equaÃ§Ã£o de recorrÃªncia desse caso Ã©
    T(n)=T(nâˆ’1)+Î˜(n)(pior caso)
    A soluÃ§Ã£o dessa recorrÃªncia Ã© T(n)=O(n2), que Ã© a mesma complexidade (no pior caso) do Bubble Sort, Insertion Sort e Selection Sort. Esse Ã© o ponto fraco do Quicksort. Embora existam vÃ¡rias versÃµes do algoritmo que tentam minimizar esse pior caso, sempre haverÃ¡ alguma configuraÃ§Ã£o na qual o Quicksort serÃ¡ O(n2).
    A complexidade no caso mÃ©dio do Quicksort Ã© O(nlogn) e Ã© um dos pontos fortes do algoritmo. O caso mÃ©dio Ã© uma medida estatÃ­stica. Isso significa que, ao executar o Quicksort, espera-se que o tempo de execuÃ§Ã£o seja O(nlogn).
</p>

```python
inicio_merge = time.time()
merge_sort(df_filter)
fim_merge = time.time()
total_time_merge =  fim_merge - inicio_merge
merge = merge_sort(df_filter)
print(" #### DEPOIS DO MERGE SORT ####")
display(merge[:50])
print(f"O fim_mergem de excuÃ§Ã£o do MERGE SORT foi {total_time_merge}")
```

```python
inicio_quick = time.time()
quick_sort(df_filter)
fim_quick = time.time()
total_time_quick =  fim_quick - inicio_quick
quick = quick_sort(df_filter)
print(" #### DEPOIS DO QUICK SORT ####")
display(quick[:50])
print(f"O tem de excuÃ§Ã£o do QUICK SORT foi {total_time_quick}")
```
